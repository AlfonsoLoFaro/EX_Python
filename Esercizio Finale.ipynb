```python
"""Si chiede poi per ogni continente di trovare il numero di casi totali avvenuti in quello stesso continente; 
si chiede di non considerare eventuali locazioni che nel dataset non appartengono ad alcun continente"""

df_pulito=covid_df.dropna(subset=["continent", "location", "total_cases"])
max_casi_per_location = df_pulito.groupby(['continent', 'location'])['total_cases'].max()
max_casi_per_location = max_casi_per_location.groupby("continent").sum()
max_casi_per_location
```

```python
"""Sempre riguardo i casi totali, si chiede di scrivere del codice che, 
date due variabili contenenti i nomi di continenti, 
ne confronti i seguenti relativi descrittori statistici: 
valore massimo,media, e percentuale rispetto al numero dei casi totali nel mondo
(calcolati anche sulle locazioni senza indicazione di continente)"""

import matplotlib.pyplot as plt

statistiche_asia_sudamerica = pd.DataFrame()

asia, s_america = "Asia", "South America"
casi_totali = df_pulito.groupby("location")["total_cases"].max().sum()

for continent in (asia, s_america):

    max_cases = [df_pulito[df_pulito['continent'] == continent].groupby("location")["total_cases"].max().sum()]
    mean = [df_pulito[df_pulito['continent'] == continent].groupby("location")["total_cases"].max().mean()]
    temp = pd.DataFrame({
                            "Continent" : continent,
                            "Max_cases" : max_cases,
                            "Mean": mean,
                            "Perc_on_total_cases" : (max_cases / casi_totali) * 100
                        }).round(2).set_index("Continent")
    statistiche_asia_sudamerica = pd.concat([statistiche_asia_sudamerica, temp])

statistiche_asia_sudamerica.Max_cases.plot(kind ="bar", color = ["red", "blue"])
statistiche_asia_sudamerica["Max_cases"].to_frame()
```

```python
"""Selezionare i dati relativi all'Italia nel 2022, e mostrare con un grafico adeguato 
l'evoluzione del casi totali rispetto alla data. 
Mostrare poi con un grafico adeguato il numero di nuovi casi rispetto alla data (filtrare i dati se necessario). 
Mostrare infine un grafico che mostra l'andamento della somma cumulativa nuovi casi del 2022, 
"""

df_italy = df_pulito[df_pulito["location"]== "Italy"]
df_italy["date"] = pd.to_datetime(df_italy["date"])
italia_nel_22 =df_italy[df_italy["date"].dt.year == 2022]
italia_nel_22["month"] = italia_nel_22["date"].dt.month

#casi totali per mese
italia_nel_22.groupby("month")["total_cases"].max().plot(kind="bar",color="red")
```

```python
#nuovi casi
italia_nel_22.groupby("month")["new_cases"].max().plot(kind="bar",color="green")

#confronto nuovi casi/casi totali
italia_nel_22.groupby("month")[["total_cases", "new_cases"]].max().plot(kind="bar")
```

```python
"""Riguardo le nazioni di Italia, Germania e Francia, mostrare in un boxplot 
la differenza tra queste nazioni riguardo il numero di pazienti in terapia intensiva (Intensive Care Unit, ICU)
da maggio 2022 (incluso) ad aprile 2023 (incluso)"""

import seaborn as sns
import matplotlib.pyplot as plt


df_ita_ger_fra = df_pulito[df_pulito['location'].isin(['Italy', 'Germany', 'France'])]
df_ita_ger_fra["date"] = pd.to_datetime(df_ita_ger_fra["date"])
df_range_time= df_ita_ger_fra[(df_ita_ger_fra['date'] >= '2022-05-01') & (df_ita_ger_fra['date'] <= '2023-04-30')]

plt.figure(figsize=(8, 4))
sns.boxplot(x='location', y='icu_patients', data= df_ita_ger_fra)
plt.title('ICU Patients (May 2022 - April 2023)')
plt.xlabel('Countries')
plt.ylabel('ICU Patients Count')
plt.show()
```
